```python
# reference
# https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html
import scipy as sc
import numpy as np
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib
from sklearn.manifold import TSNE
import warnings
from matplotlib import pyplot as plt
from sklearn.mixture import GaussianMixture as GMM
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=2)
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
def original_labels(datapoints,labels_origin,numclusters=3):
# print (datapoints)
    # svd = TruncatedSVD(n_components=2)
    tnsepoints = datapoints
    clusters = {}
    predicted_labels = []
    for i in range(len(labels_origin)):
        predicted_labels.append(labels_origin[i])
        if(labels_origin[i] not in clusters):
            clusters[labels_origin[i]] = []
        clusters[labels_origin[i]].append(i)
    for i in clusters.keys():
        clusters[i] = tnsepoints[clusters[i]]
    # print (clusters)
    plotgraph(clusters,"Original ")
    # print ("Clusters using Kmeans: %d"%(len(set(kmeans.labels_))))
```

```python
def completefun(datapoints,numclusters=3):
# print (datapoints)
    kmeans = KMeans(n_clusters=numclusters, random_state=0).fit(datapoints)
    # svd = TruncatedSVD(n_components=2)
    tnsepoints = datapoints
    clusters = {}
    predicted_labels = []
    for i in range(len(kmeans.labels_)):
        predicted_labels.append(kmeans.labels_[i])
        if(kmeans.labels_[i] not in clusters):
            clusters[kmeans.labels_[i]] = []
        clusters[kmeans.labels_[i]].append(i)
    for i in clusters.keys():
        clusters[i] = tnsepoints[clusters[i]]
    # print (clusters)
    plotgraph(clusters,"Kmeans ")
    return kmeans.labels_
    # print ("Clusters using Kmeans: %d"%(len(set(kmeans.labels_))))
```

```python
def completefun_gmm(datapoints,numclusters=3):
# print (datapoints)
    a,b = build_gmm(datapoints,numclusters)
    # svd = TruncatedSVD(n_components=2)
    gmmlabels = b
    tnsepoints = datapoints
    clusters = {}
    predicted_labels = []
    for i in range(len(gmmlabels)):
        predicted_labels.append(gmmlabels[i])
        if(gmmlabels[i] not in clusters):
            clusters[gmmlabels[i]] = []
        clusters[gmmlabels[i]].append(i)
    for i in clusters.keys():
        clusters[i] = tnsepoints[clusters[i]]
    # print (clusters)
    plotgraph(clusters,"GMM ")
    return gmmlabels
    # print ("Clusters using Kmeans: %d"%(len(set(kmeans.labels_))))
```

```python
def plotgraph(resultlist,title=""):
    '''
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    '''
    # plt.figure()
    plt.figure(num=None, figsize=(10,10), dpi=80, facecolor='w', edgecolor='k')
    for i in sorted(resultlist.keys()):
        plt.scatter(resultlist[i].T[0],resultlist[i].T[1],label='Cluster ' + str(i))
    #plt.show()
    plt.legend(loc='best')
    plt.title(title)
    plt.show()
```

```python
def build_gmm(data,K): 
        """
        Build gmm from data
        """

        gmm = GMM(n_components=K, covariance_type='diag',random_state=0)
        labels_gmm = gmm.fit_predict(data)
        
        return gmm.means_,labels_gmm
```

```python
mu = [0.0,3,9]
sigma = [.9,.9,.9]
samplesize = 1000
dists = []
dimension = 2
dists.append(np.random.normal(mu[0], sigma[0], size=[dimension,samplesize]).T)
dists.append(np.random.normal(mu[1], sigma[1], size=[dimension,samplesize]).T)
dists.append(np.random.normal(mu[2], sigma[2], size=[dimension,samplesize]).T)
truelabels = []
for i in range(samplesize):
    truelabels.append(0)
for i in range(samplesize):
    truelabels.append(1)
for i in range(samplesize):
    truelabels.append(2)
completedata = []
for i in range(len(dists)):
    completedata.extend(dists[i])
```

```python
truelabels = np.array(truelabels)
completedata = np.array(completedata)
```

```python
X = completefun_gmm(completedata,3)
```

```python
X = completedata
values = []
GG = []
for n_clusters in range(2,13):
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)
    values.append(silhouette_avg)
    GG.append(n_clusters)
plt.plot(GG,values)
```
